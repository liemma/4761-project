```python
import pandas as pd

# Metadata of the original dataset
merfish_metadata = abc_cache.get_metadata_dataframe(
    directory="Zhuang-ABCA-1",
    file_name="cell_metadata",
    dtype={"cell_label": str},
)
merfish_metadata = merfish_metadata.set_index("cell_label")

# CCF-aligned metadata of the same dataset
ccf_metadata = abc_cache.get_metadata_dataframe(
    directory="Zhuang-ABCA-1-CCF",
    file_name="cell_metadata",
    dtype={"cell_label": str},
)
ccf_metadata = ccf_metadata.set_index("cell_label")

# Add the subregion information from the CCF aligned metadata to the original metadata
if "parcellation_substructure_label" in ccf_metadata.columns:
    merfish_metadata["subregion"] = ccf_metadata["parcellation_substructure_label"]

# Count cells per brain section (slice)
section_col = "brain_section_label" if "brain_section_label" in merfish_metadata.columns else "section_label"
slice_counts = merfish_metadata[section_col].value_counts().sort_index()
print("Slices (top 20):")
print(slice_counts.head(20))
```
